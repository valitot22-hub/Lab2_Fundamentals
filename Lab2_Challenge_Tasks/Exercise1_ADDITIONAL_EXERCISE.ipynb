```python
# [CELL 1] Notebook "ADDITIONAL EXERCISE"

target_name = "GOLISAO"
student_id = "TUPM-25-0803"  
seed_num = student_id[-1]        
redacted_name = ""

vowels = "AEIOUaeiou"

for char in target_name:
    if char in vowels:
        redacted_name += seed_num
    else:
        redacted_name += char

print(f"Original Name: {target_name}")
print(f"Seed Number:   {seed_num}")
print(f"Redacted Name: {redacted_name}")
```
